```python
import openpyxl  # подключаем библиотеку openpyxl для работы с Excel (.xlsx)
import matplotlib.pyplot as plt  # подключаем модуль matplotlib для построения графиков
```

```python
wb = openpyxl.load_workbook("students.xlsx")  # открываем Excel-файл (файл должен лежать рядом с ноутбуком)
print("Листы в файле:", wb.sheetnames)  # выводим список всех листов, чтобы проверить их названия
sheet = wb["Оценки"]  # выбираем лист с названием "Оценки"
```

```python
fio_list = []  # создаём пустой список для хранения ФИО студентов
math_scores = []  # создаём пустой список для хранения оценок по математике
physics_scores = []  # создаём пустой список для хранения оценок по физике
informatics_scores = []  # создаём пустой список для хранения оценок по информатике
```

```python
for row in sheet.iter_rows(min_row=2, values_only=True):  # цикл по строкам листа, начиная со 2-й (первая строка — заголовки)
    fio, math, physics, informatics = row  # раскладываем каждую строку на 4 переменные: ФИО, математика, физика, информатика
    fio_list.append(fio)  # добавляем ФИО в список fio_list
    math_scores.append(math)  # добавляем оценку по математике в список math_scores
    physics_scores.append(physics)  # добавляем оценку по физике в список physics_scores
    informatics_scores.append(informatics)  # добавляем оценку по информатике в список informatics_scores
```

```python
def percent_below_6(scores):  # определяем функцию для подсчёта процента оценок ниже 6
    return sum(1 for v in scores if v < 6) / len(scores) * 100  # считаем количество оценок <6, делим на общее число и переводим в проценты

math_percent = percent_below_6(math_scores)  # считаем процент оценок <6 по математике
physics_percent = percent_below_6(physics_scores)  # считаем процент оценок <6 по физике
informatics_percent = percent_below_6(informatics_scores)  # считаем процент оценок <6 по информатике
```

```python
subjects = ["Математика", "Физика", "Информатика"]  # список названий предметов
percents = [math_percent, physics_percent, informatics_percent]  # список рассчитанных процентов

plt.bar(subjects, percents, color=["blue", "green", "orange"])  # строим столбчатую диаграмму с цветами для каждого предмета
plt.title("Процент учащихся с оценкой ниже 6")  # добавляем заголовок диаграммы
plt.ylabel("Процент (%)")  # подписываем ось Y
plt.legend(["<6"])  # добавляем легенду "<6"
plt.show()  # показываем график
```

```python
wb_result = openpyxl.Workbook()  # создаём новый Excel-файл (рабочую книгу)

sheet_result = wb_result.active  # выбираем активный (первый) лист в книге
sheet_result.title = "Результаты"  # задаём название листа "Результаты"

sheet_result["A1"] = "Предмет"  # в ячейку A1 записываем заголовок "Предмет"
sheet_result["B1"] = "Процент < 6"  # в ячейку B1 записываем заголовок "Процент < 6"

sheet_result.append(["Математика", math_percent])  # добавляем строку с предметом "Математика" и рассчитанным процентом
sheet_result.append(["Физика", physics_percent])  # добавляем строку с предметом "Физика" и рассчитанным процентом
sheet_result.append(["Информатика", informatics_percent])  # добавляем строку с предметом "Информатика" и рассчитанным процентом

wb_result.save("students_result.xlsx")  # сохраняем книгу в файл "students_result.xlsx"
print("Файл students_result.xlsx успешно создан!")  # выводим сообщение об успешном создании файла
```
