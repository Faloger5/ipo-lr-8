Работа с интерактивными блокнотами и файлами

```python
import openpyxl
import matplotlib.pyplot as plt

wb = openpyxl.load_workbook("students.xlsx")
sheet = wb["Оценки"]

fio_list, math_scores, physics_scores, informatics_scores = [], [], [], []
for row in sheet.iter_rows(min_row=2, values_only=True):
    fio, math, physics, informatics = row
    fio_list.append(fio)
    math_scores.append(math)
    physics_scores.append(physics)
    informatics_scores.append(informatics)

def percent_below_6(scores):
    return sum(1 for v in scores if v < 6) / len(scores) * 100

math_percent = percent_below_6(math_scores)
physics_percent = percent_below_6(physics_scores)
informatics_percent = percent_below_6(informatics_scores)

subjects = ["Математика", "Физика", "Информатика"]
percents = [math_percent, physics_percent, informatics_percent]
plt.bar(subjects, percents, color=["blue", "green", "orange"])
plt.title("Процент учащихся с оценкой ниже 6")
plt.ylabel("Процент (%)")
plt.legend(["<6"])
plt.show()

wb_result = openpyxl.Workbook()
sheet_result = wb_result.active
sheet_result.title = "Результаты"
sheet_result["A1"] = "Предмет"
sheet_result["B1"] = "Процент < 6"
sheet_result.append(["Математика", math_percent])
sheet_result.append(["Физика", physics_percent])
sheet_result.append(["Информатика", informatics_percent])
wb_result.save("students_result.xlsx")
print("Файл students_result.xlsx успешно создан!")
```
